```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
plt.title('Duration of songs of different Genre')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel('Duration in milli seconds')
plt.ylabel('Genres')
```

```python
df_popularity = df_genre.sort_values('popularity', ascending=False)
df_popularity.head(5)
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize=(10, 5))
famous = df_genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data = famous).set(title='Top 5 Genre by popularity')
```
